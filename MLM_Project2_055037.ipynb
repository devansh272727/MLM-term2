#### **5.5. Machine Learning Models**

-   **Logistic Regression (LR)**: Logistic regression is a classification algorithm that models the probability of a binary outcome based on one or more predictor variables using a logistic function.
    
-   **Support Vector Machines (SVM)**: SVM finds a hyperplane that best separates the data into classes by maximizing the margin between the closest points of the classes (support vectors).
    
-   **Stochastic Gradient Descent (SGD)**: An iterative optimization algorithm used for minimizing loss functions in linear classifiers and regressors, particularly useful for large datasets.
    
-   **Decision Trees**: A tree-based algorithm where each internal node represents a decision on a feature, each leaf node represents an outcome, and paths from root to leaf represent classification rules.
    
-   **K-Nearest Neighbors (KNN)**: A lazy learning algorithm that classifies a data point based on the majority class of its k-nearest neighbors in the feature space.
    
-   **Naive Bayes (NB)**: A probabilistic classifier based on Bayes' theorem, assuming strong independence between features.
    
-   **Bagging (Bootstrap Aggregating)**: Combines predictions from multiple models (e.g., Random Forest) to reduce variance and improve accuracy.
    
-   **Boosting**: Combines weak learners sequentially to form a strong learner by minimizing errors iteratively (e.g., Extreme Gradient Boosting - XGBoost).

```python
# Confidence Interval for the Mean
def confidence_interval(data, confidence=0.95):
    mean = data.mean()
    std = data.std()
    n = len(data)
    z = stats.norm.ppf(1 - (1 - confidence) / 2)  # z-score for 95% CI
    margin_of_error = z * (std / np.sqrt(n))
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error
    return (lower_bound, upper_bound)

ci_results = {}
for col in dp37_non_cat_vars:
    ci_results[col] = confidence_interval(dp37_sample[col])


print("\nConfidence Intervals (95%):")
for col, ci in ci_results.items():
    print(f"{col}: {ci[0]:.2f} to {ci[1]:.2f}")
```

## **Inferential Statistics**

```python
# Test of Homogeneity {Chi-sq} on categorical data

dp37_contingency_table = pd.crosstab(dp37_sample['Import_Export'], dp37_sample['Category'])

# Perform the Chi-square test
chi2, p, dof, expected = chi2_contingency(dp37_contingency_table)

print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of freedom: {dof}")
print("Expected frequencies:")
print(expected)

# Interpret the results
dp37_alpha = 0.05  # Significance level

if p < dp37_alpha:
    print("Reject the null hypothesis: There is a significant association between the two categorical variables.")
else:
    print("There is no significant association between the two categorical variables.")
```

```python
# Test of Normality {Shapiro-Wilk | Kolmogorov-Smirnov | Anderson-Darling | Jarque-Bera

def normality_tests(data):

    dp37_results = {}

    # Shapiro-Wilk test
    shapiro_statistic, shapiro_pvalue = shapiro(data)
    dp37_results["Shapiro-Wilk"] = {"statistic": shapiro_statistic, "p-value": shapiro_pvalue}

    # Kolmogorov-Smirnov test (against a normal distribution)
    ks_statistic, ks_pvalue = kstest(data, 'norm')
    dp37_results["Kolmogorov-Smirnov"] = {"statistic": ks_statistic, "p-value": ks_pvalue}

    # Anderson-Darling test
    anderson_result = anderson(data)
    dp37_results["Anderson-Darling"] = {
        "statistic": anderson_result.statistic,
        "critical_values": anderson_result.critical_values,
        "significance_level": anderson_result.significance_level
    }

    # Jarque-Bera test
    jb_statistic, jb_pvalue = jarque_bera(data)
    dp37_results["Jarque-Bera"] = {"statistic": jb_statistic, "p-value": jb_pvalue}

    return dp37_results
```

```python
from scipy import stats

# Test of Correlation (T-test)
def t_test_correlation(data1, data2):
    t_statistic, p_value = stats.ttest_ind(data1, data2)
    return {"t-statistic": t_statistic, "p-value": p_value}

for col in dp37_non_cat_vars:
    print(f"Normality tests for {col}:")
    normality_test_results = normality_tests(dp37_sample[col])
    for test_name, result in normality_test_results.items():
        print(f"  {test_name}: {result}")

    # Example of T-test
    if len(dp37_non_cat_vars) > 1:
        for i in range(len(dp37_non_cat_vars)):
            for j in range(i+1, len(dp37_non_cat_vars)):
                col1, col2 = dp37_non_cat_vars[i], dp37_non_cat_vars[j]
                t_test_result = t_test_correlation(dp37_sample[col1], dp37_sample[col2])
                print(f"T-test Correlation between {col1} and {col2}: {t_test_result}")
    print()
    print('*'*20)
    print()
```

# Model Run Statistics

## Time Taken

```python
import time
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

def time_model_run(model, X_train, y_train):
    start_time = time.time()
    model.fit(X_train, y_train)
    end_time = time.time()
    return end_time - start_time

# Logistic Regression
log_reg_model = LogisticRegression(random_state=55037)
log_reg_time = time_model_run(log_reg_model, dp37_X_train, dp37_y_train)
print(f"Logistic Regression runtime: {log_reg_time:.4f} seconds")

# Support Vector Machine (SVM)
svm_model = SVC(random_state=55037)
svm_time = time_model_run(svm_model, dp37_X_train, dp37_y_train)
print(f"SVM runtime: {svm_time:.4f} seconds")

# Stochastic Gradient Descent (SGD)
sgd_model = SGDClassifier(random_state=55037)
sgd_time = time_model_run(sgd_model, dp37_X_train, dp37_y_train)
print(f"SGD runtime: {sgd_time:.4f} seconds")

# Decision Tree
dt_model = DecisionTreeClassifier(random_state=55037)
dt_time = time_model_run(dt_model, dp37_X_train, dp37_y_train)
print(f"Decision Tree runtime: {dt_time:.4f} seconds")

# K-Nearest Neighbors (KNN)
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_time = time_model_run(knn_model, dp37_X_train, dp37_y_train)
print(f"KNN runtime: {knn_time:.4f} seconds")

# Naive Bayes
nb_model = GaussianNB()
nb_time = time_model_run(nb_model, dp37_X_train, dp37_y_train)
print(f"Naive Bayes runtime: {nb_time:.4f} seconds")

# Random Forest
rf_model = RandomForestClassifier(random_state=55037)
rf_time = time_model_run(rf_model, dp37_X_train, dp37_y_train)
print(f"Random Forest runtime: {rf_time:.4f} seconds")

# Extreme Gradient Boosting (XGBoost)
xgb_model = XGBClassifier(random_state=55037)
xgb_time = time_model_run(xgb_model, dp37_X_train, dp37_y_train)
print(f"XGBoost runtime: {xgb_time:.4f} seconds")
```

**Observations:**

-   **Fastest:** KNN and Naive Bayes are the fastest algorithms, with runtimes under 0.01 seconds.
-   **Moderate Speed:** Logistic Regression, SGD, and Decision Tree have moderate runtimes, ranging from 0.0101 to 0.0440 seconds.
-   **Relatively Slow:** XGBoost has a relatively slow runtime of 0.2831 seconds.
-   **Slowest:** SVM and Random Forest are the slowest algorithms, with runtimes of 0.5676 seconds and 1.1377 seconds, respectively.

**Insights:**

-   **KNN and Naive Bayes** are ideal for applications where speed is critical, such as real-time predictions or when computational resources are limited.
-   **Logistic Regression, SGD, and Decision Tree** offer a good balance of speed and complexity, making them suitable for a wide range of applications where moderate computational resources are available.
-   **XGBoost** provides a balance between speed and accuracy, but requires more computational resources than the faster algorithms.
-   **SVM and Random Forest** are generally slower algorithms, requiring more computational resources. They might be more suitable for tasks where accuracy is paramount and computational resources are not a major constraint, or when the complexity of the model is expected to improve accuracy significantly.


## Memory Utilized

```python
import tracemalloc
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

def get_model_memory_usage(model, X, y):
    tracemalloc.start()
    model.fit(X, y)
    current, peak = tracemalloc.get_traced_memory()
    tracemalloc.stop()
    return peak / (1024 * 1024)  # Convert bytes to MB

# Logistic Regression
log_reg_model = LogisticRegression(random_state=55037)
log_reg_memory = get_model_memory_usage(log_reg_model, dp37_X_train, dp37_y_train)
print(f"Logistic Regression memory usage: {log_reg_memory:.2f} MB")

# Support Vector Machine (SVM)
svm_model = SVC(random_state=55037)
svm_memory = get_model_memory_usage(svm_model, dp37_X_train, dp37_y_train)
print(f"SVM memory usage: {svm_memory:.2f} MB")

# Stochastic Gradient Descent (SGD)
sgd_model = SGDClassifier(random_state=55037)
sgd_memory = get_model_memory_usage(sgd_model, dp37_X_train, dp37_y_train)
print(f"SGD memory usage: {sgd_memory:.2f} MB")

# Decision Tree
dt_model = DecisionTreeClassifier(random_state=55037)
dt_memory = get_model_memory_usage(dt_model, dp37_X_train, dp37_y_train)
print(f"Decision Tree memory usage: {dt_memory:.2f} MB")

# K-Nearest Neighbors (KNN)
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_memory = get_model_memory_usage(knn_model, dp37_X_train, dp37_y_train)
print(f"KNN memory usage: {knn_memory:.2f} MB")

# Naive Bayes
nb_model = GaussianNB()
nb_memory = get_model_memory_usage(nb_model, dp37_X_train, dp37_y_train)
print(f"Naive Bayes memory usage: {nb_memory:.2f} MB")

# Random Forest
rf_model = RandomForestClassifier(random_state=55037)
rf_memory = get_model_memory_usage(rf_model, dp37_X_train, dp37_y_train)
print(f"Random Forest memory usage: {rf_memory:.2f} MB")

# Extreme Gradient Boosting (XGBoost)
xgb_model = XGBClassifier(random_state=55037)
xgb_memory = get_model_memory_usage(xgb_model, dp37_X_train, dp37_y_train)
print(f"XGBoost memory usage: {xgb_memory:.2f} MB")
```

**Observations:**

-   **Highest:** XGBoost has the highest memory usage, requiring 0.88 MB.
-   **Relatively High:** Random Forest has relatively high memory usage, requiring 0.49 MB.
-   **Moderate:** Logistic Regression (0.34 MB), SVM (0.27 MB), Decision Tree (0.22 MB), and KNN (0.22 MB) have moderate memory usage.
-   **Low:** SGD has low memory usage, requiring 0.17 MB.
-   **Lowest:** Naive Bayes has the lowest memory usage, requiring only 0.11 MB.

**Insights:**

-   **XGBoost** has the highest memory usage, requiring careful resource management when working with large datasets or on systems with limited memory.
-   **Random Forest** also has relatively high memory usage, which could be a concern for applications with limited memory resources.
-   **Logistic Regression, SVM, Decision Tree, and KNN** have moderate memory requirements, making them suitable for most use cases with average memory capacity.
-   **SGD** has low memory usage, making it suitable for resource-constrained environments or when working with very large datasets.
-   **Naive Bayes** has the lowest memory footprint, making it ideal for memory-sensitive tasks or applications where memory usage is a critical factor.
